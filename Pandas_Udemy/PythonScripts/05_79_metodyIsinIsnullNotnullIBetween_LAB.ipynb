```python
"""1. Zaimportuj moduł pandas oraz numpy i nadaj im standardowe aliasy. Do zmiennej fuel wczytaj zawartość pliku
fuel.csv. Podczas wczytywania skorzystaj z dodatkowego argumentu low_memory=False, pobierz tylko
następujące kolumny: 'Vehicle ID','Year','Make','Model','Class','Fuel Type','Combined MPG (FT1)','Start Stop
Technology'. Wyświetl nagłówek tak utworzonego Data Frame"""

import pandas as pd
import numpy as np
fuel = pd.read_csv('/Users/kamilw/Desktop/Moje/PythonLearing/Pandas_Udemy/course-files/fuel.csv',low_memory=False,usecols=['Vehicle ID','Year','Make','Model','Class','Fuel Type','Combined MPG (FT1)','Start Stop Technology'],index_col='Vehicle ID')
fuel.head()
```

```python
"""2. Utwórz zmienną listOfCars typu lista i wpisz do niej napisy: 'Renault', 'Toyota','Ford'. Utwórz zmienną
isInList i korzystając z metody pokazanej na lekcji przypisz do niej serię z wartościami True/False - True,
jeżeli w kolumnie "Make" jest wartość z listy listOfCars, False - w przeciwnym razie. Wyświetl nagłówek tej
serii"""

listOfCars = ['Renault','Toyota','Ford']
isInList = fuel['Make'].isin(listOfCars)
isInList.head()
```

```python
"""3. Wyświetl z fuel, te samochody, które są na liście listOfCars. Wykorzystaj w tym celu zmienną isInList"""

fuel[isInList].head()
```

```python
"""4. Utwórz zmienną isNoStartStopDefined i korzystając z metody pokazanej na lekcji przypisz do niej serię z
wartościami True/False - True, jeżeli w kolumnie "Start Stop Technology" jest wartość NaN, False - w
przeciwnym razie. Wyświetl nagłówek tej serii"""

isNoStartStopDefined = fuel['Start Stop Technology'].isnull()
isNoStartStopDefined.head()
```

```python
"""5. Wyświetl z fuel te samochody, o których nie wiadomo czy mają system Start/Stop. Wykorzystaj w tym celu
zmienną isNoStartStopDefined"""

fuel[isNoStartStopDefined].head()
```

```python
"""6. Utwórz zmienną isStartStopDefined i korzystając z metody pokazanej na lekcji przypisz do niej serię z
wartościami True/False - True, jeżeli w kolumnie "Start Stop Technology" jest zdefiniowna jakaś wartość,
False - w przeciwnym razie. Wyświetl nagłówek tej serii"""

isStartStopDefined = fuel['Start Stop Technology'].notnull()
isStartStopDefined.head()
```

```python
"""7. Wyświetl z fuel te samochody, o których wiadomo czy mają system Start/Stop. Wykorzystaj w tym celu
zmienną isStartStopDefined"""

fuel[isStartStopDefined].head()
```

```python
"""8. Utwórz zmienną mpg50_60 i korzystając z metody pokazanej na lekcji przypisz do niej serię z wartościami
True/False - True, jeżeli w kolumnie "Combined MPG (FT1)" jest wartość >=50 i <=60, False - w przeciwnym
razie. Wyświetl nagłówek tej serii"""

mpg50_60 = fuel['Combined MPG (FT1)'].between(50,60)
mpg50_60.head()
```

```python
"""9. Wyświetl z fuel te samochody, które na jednym galonie paliwa przejadą od 50-60 mil. Wykorzystaj w tym celu
zmienną mpg50_60"""

fuel[mpg50_60].head()
```

```python
"""10. Korzystając ze zdefiniowanych wcześniej zmiennych wyświetl komunikat podobny do poniższego
There are 4376 cars within the list ['Renault', 'Toyota', 'Ford']
There are 31374 cars where Start/Stop technology is not determined
There are 310 cars where Start/Stop technology is determined
There are 13 cares with MPG value between 50 and 60"""

print('There are ',len(fuel[isInList]),'cars within the list ',listOfCars)
print('There are %s cars where Start/Stop technology is not determined' % len(fuel[isNoStartStopDefined]))
print('There are %s cars where Start/Stop technology is determined' % len(fuel[isStartStopDefined]))
print('There are %s cars with MPG value between 50 and 60' % len(fuel[mpg50_60]))
```
