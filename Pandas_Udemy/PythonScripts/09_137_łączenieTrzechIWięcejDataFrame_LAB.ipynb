```python
"""1. Zaimportuj moduł pandas i numpy nadaj im standardowe aliasy"""

import pandas as pd
import numpy as np
```

```python
"""2. Do zmiennej suppliers zaimportuj zawartość pliku suppliers.csv. Pobierz tylko kolumny SupplierID,
CompanyName, City"""

suppliers = pd.read_csv("/Users/kamilw/Desktop/Moje/PythonLearing/Pandas_Udemy/course-files/northwind-mongo-master/suppliers.csv",
                       usecols=['SupplierID','CompanyName','City'])
```

```python
"""3. Do zmiennej products zaimportuj zawartość pliku products.csv. Pobierz tylko kolumny ProductID,
ProductName, SupplierID, CategoryID"""

products = pd.read_csv("/Users/kamilw/Desktop/Moje/PythonLearing/Pandas_Udemy/course-files/northwind-mongo-master/products.csv",
                      usecols=['ProductName','SupplierID','CategoryID'])
```

```python
"""4. Do zmiennej categories zaimportuj zawartość pliku categories1.csv. Pobierz tylko kolumny CategoryID i
CategoryName. Kolumna CategoryID ma być indeksem."""

categories = pd.read_csv("/Users/kamilw/Desktop/Moje/PythonLearing/Pandas_Udemy/course-files/northwind-mongo-master/categories1.csv",
                        usecols=['CategoryID','CategoryName'],index_col='CategoryID')
```

```python
"""5. Połącz products i categories korzystając z CategoryID. Wynik zapisz w nowej zmiennej prod_cat"""

prod_cat = products.merge(categories,on='CategoryID')
```

```python
prod_cat.head()
```

```python
"""6. Połącz prod_cat z suppliers korzytając z SupplierID. Wynik zapisz w nowej zmiennej prod_cat_sup"""

prod_cat_sup = prod_cat.merge(suppliers,on='SupplierID')
```

```python
prod_cat_sup.head()
```

```python
"""7. Wykonaj złączenia opisane w krokach (5) i (6), ale bez stosowania pomocniczych zmiennych"""

products.merge(categories,on='CategoryID').merge(suppliers,on='SupplierID')
```
