```python
"""1. Zaimportuj moduł pandas oraz numpy i nadaj im standardowe aliasy. Do zmiennej animals wczytaj
zawartość pliku sleep_time.csv. Pobierz wszystkie kolumny, a kolumnę ID wybierz jako indeks. Podczas
wczytywania zamień wszystkie wartości NaN na 0. Wyświetl nagłówek zaimportowanego pliku"""

import pandas as pd
import numpy as np
animals = pd.read_csv("/Users/kamilw/Desktop/Moje/PythonLearing/Pandas_Udemy/course-files/sleep_time.csv",index_col='ID').fillna(value=0)
animals.head()
```

```python
"""2. 2. W kolumnie vore znajduje się informacja o tym czy zwierzę jest roślinożerne (herbi), mięsożerne (carni) i
wszystkożerne (omni). Korzystając z metody where wybierz wiersze opisujące zwierzęta roślinożerne.
Wynik zapisz w zmiennej herbi"""

herbiVore = animals['vore'] == "herbi"
herbi = animals.where(herbiVore)
herbi.head()
```

```python
"""3. Ile wierszy jest w zmiennej herbi?"""

len(herbi)
```

```python
"""4. Ile wierszy jest w zmiennej herbi po usunięciu wartości NaN?"""

len(herbi.dropna())
```

```python
"""5. Utwórz zmienną carni i zapisz w niej wynik filtrowania obiektu animals tak, że pobrane zostaną tylko
wiersze zwierząt mięsożernych. Wyświetl informację o ilości wierszy w zmiennej carni oraz o ilości wierszy
po usunięciu wartości NaN"""

carni = animals.where(animals['vore'] == 'carni')
print(len(carni))
print(len(carni.dropna()))
```

```python
"""6. Utwórz zmienną omni i zapisz w niej wynik filtrowania obiektu animals tak, że pobrane zotaną tylko
wiersze zwierząt wszystkożernych. Wyświetl informację o ilości wierszy w zmiennej omni oraz o ilości
wierszy po usunięciu wartości NaN."""

omni = animals.where(animals['vore'] == 'omni')
print(len(omni))
print(len(omni.dropna()))
```

```python
"""7. Wyświetl średni czas snu (kolumna sleep_total) dla zwierząt roślino-, mieso- i wszystkożernych. Czy
można próbować postawić hipotezę, że rodzaj pokarmu wpływa na czas snu?"""

print('carni mean sleep time:', carni['sleep_total'].mean())
print('herbi mean sleep:', herbi['sleep_total'].mean())
print('omni mean sleep:', omni['sleep_total'].mean())
```

```python
"""8. Korzystając z polecenia query wybierz te wiersze, które opisują zwierzęta śpiące więcej niż 14 godzin na
dobę (kolumna sleep_total)"""

animals.query('sleep_total > 14').head()
```

```python
"""9. Do warunku z poprzedniego zadania dodaj: rodzaj zwierzęcia (kolumna vore) ma być roślinożerny
(wartość herbi)"""

animals.query('sleep_total > 14 and vore == "herbi"').head()
```

```python
"""10. Do warunku z poprzedniego zadania dodaj: waga ciała (kolumna bodywt) ma być większa niż 1"""

animals.query('sleep_total > 14 and vore == "herbi" and bodywt > 1').head()
```
