```python
"""1. Zaimportuj moduł pandas i numpy, nadaj im standardowe aliasy. Do zmiennej professions wczytaj
zawartość pliku Prestige.csv. Wyświetl nagłówek obiektu data frame"""

import pandas as pd
import numpy as np
professions = pd.read_csv('/Users/kamilw/Desktop/Moje/PythonLearing/Pandas_Udemy/course-files/Prestige.csv')
professions.head()
```

```python
"""2. Zdefiniuj słownik w zmiennej dict, który ma klucze takie, jak nazwy kolumn w professions. Wpisz do
wartości dla klucza name - data scientist, pozostałe wartości zdefiniuj jako NaN. Wyświetl obiekt w celu
werryfikacji."""

dict = {'name':'data scientist','education':np.NaN,'income':np.NaN,'women':np.NaN,'prestige':np.NaN,'census':np.NaN,'type':np.NaN}
dict
```

```python
"""3. Dodaj do profession nowy wiersz zdefiniowany przez słownik dict. W celu weryfikacji wyświetl 5 ostatnich
wartości z profession"""

professions.append(dict, ignore_index=True).tail(5)
```

```python
"""4. Odszukaj w profession wiersza, w którym name to taxi.drivers. Zapisz go do zmiennej taxi i wyświetl."""

taxi = professions[professions["name"] == 'taxi.drivers']
taxi
```

```python
"""5. Usuń z profession wiersz z indekstem 98 (to właśnie taxi.drivers). Wyświetl 5 ostatnich wartości z
profession"""

professions.drop(98,inplace=True)
professions.tail(10)
```

```python
"""6. Dodaj wiersz znajdujący się w taxi do profession, w taki sposób, aby wiersz na stałe znajdował sie w
zmiennej profession. Wyświetl ostatnich 5 wierszy profession w celu weryfikacji"""

professions = professions.append(taxi)
professions.tail()
```

```python
"""7. Usuń z profession kolumnę census"""

del professions['census']
professions.tail()
```

```python
"""8.  Usuń z profession kolumnę type. Wykorzystaj do tego inną metodę niż wykorzystana w poprzednim
zadaniu"""

professions.drop(columns='type', inplace=True)
professions.tail()
```
