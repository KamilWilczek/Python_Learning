```python
"""1. Zaimportuj moduł pandas oraz numpy i nadaj im standardowe aliasy. Do zmiennej fortune wczytaj zawartość
pliku Fortune_500_2017.csv. Pobierz tylko następujące kolumny:
'Rank','Title','Employees','Profits','Assets'. Jako indeksu użyj kolumny 'Rank'. Wyświetl nagłówek tak
utworzonego Data Frame"""

import pandas as pd
import numpy as np
fortune = pd.read_csv('/Users/kamilw/Desktop/Moje/PythonLearing/Pandas_Udemy/course-files/Fortune_500_2017.csv',usecols=['Rank','Title','Employees','Profits','Assets'],index_col='Rank')
fortune.head()
```

```python
"""2. Wylicz ranking tak, aby najniższe wartości były przyznawane firmom z największą ilośćią pracowników
(kolumna Employees). Wyświetl nagłówek wyznaczonej serii danych"""

fortune['Employees'].rank(ascending=False).head()
```

```python
"""3. Dodaj do zmiennej frame kolumnę nazwaną 'RankByEmployee', zawierającą powyżej wyznaczony ranking"""

fortune['RankByEmployee'] = fortune['Employees'].rank(ascending=False)
fortune.head(10)
```

```python
"""4. Wylicz ranking tak, aby najniższe wartości były przyznawane firmom z największym zyskiem (kolumna
Profits). Wyświetl nagłówek wyznaczonej serii danych"""

fortune['Profits'].rank(ascending=False).head()
```

```python
"""5. Dodaj do zmiennej frame kolumnę nazwaną 'RankByProfits', zawierającą powyżej wyznaczony ranking"""

fortune['RankByProfits'] = fortune['Profits'].rank(ascending=False)
fortune.head(10)
```

```python
fortune.sort_values(by='Profits',ascending=False).head(10)
```

```python
"""6. Wyświetl 3 firmy, które mają największy zasób środków trwałych (kolumna Asset)"""

fortune.nlargest(n=3, columns='Assets')
```

```python
"""7. Wyświetl 3 firmy, które mają najmniejszy zasób środków trwałych (kolumna Asset)"""

fortune.nsmallest(n=3, columns='Assets')
```
