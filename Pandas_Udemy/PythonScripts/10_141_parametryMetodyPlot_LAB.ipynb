```python
"""3. Wyświetl wykres dla nasa_by_day w domyślnej postac"""

nasa_by_day.plot()
```

```python
"""4. Wyświetl wykres w rozmiarze 15" na 5". Kolejne wykresy też mają używać tego rozmiaru"""

nasa_by_day.plot(figsize=(15,5))
```

```python
"""5. Do wykresu dodaj tytuł "NASA Facebook activity". Kolejne wykresy też mają używać tego rozmiaru"""

nasa_by_day.plot(figsize=(15,5),title='NASA Facebook activity')
```

```python
"""6. Wyświetl wykres tak aby każda seria danych została pokazana na oddzielnym wykresie. Ta zmiana nie ma
obowiązywać w kolejnych wykresach."""

nasa_by_day.plot(figsize=(15,5),title='NASA Facebook activity',subplots=True)
```

```python
"""7. Ponieważ dane prezentowane na wykresie znacznie różnią się od siebie (część jest bardzo duża a pozostałe
są małe), to zmień skalę osi y na logarytmiczną. Ta zmiana nie ma obowiązywać w kolejnych wykresach"""

nasa_by_day.plot(figsize=(15,5),title='NASA Facebook activity',logy=True)
```

```python
"""8. Wyświetlając wykres spraw, aby na osi x były wyświetlane wszystkie numery dni od 1 do 31 włącznie."""

nasa_by_day[:31].plot(figsize=(15,5),title='NASA Facebook activity')
```

```python
"""9. Wyświetlając wykres spraw, aby na osi x były wyświetlane wszystkie numery dni od 1 do 31 włacznie. Wartości
te mają być wyświetlane pod kątem 45 stopni."""

nasa_by_day.plot(figsize=(15,5),title='NASA Facebook activity',xticks=range(1,32),rot=45)
```

```python
"""10. Wyświetl wykres dla pierwszych dziesięciu dni. Dodaj parametr, który spowoduje dodatkowo wyświetlenie pod
wykresem tabeli z dokładnymi wartościami prezenowanymi również na wykresie."""

nasa_by_day[:10].plot(figsize=(15,5),title='NASA Facebook activity',rot=45,table=True)
```
