```python
# sytuacje beznadziejne kiedy nie można używać gotowych metod
# apply
def getComment(row):
    reporterType = row['Reporter Type']
    cost = float(row['Total Operating Expenses'].replace('$',''))
    
    if(cost >200000):
        comment = "CLASS A"
    else:
        comment = 'CLASS B'
    return ( reporterType+'/'+comment)
```

```python
frame.apply(getComment,axis=1).head()
```
