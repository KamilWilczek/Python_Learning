```python
"""1. Zaimportuj moduł pandas i numpy nadaj im standardowe aliasy"""

import pandas as pd
import numpy as np
```

```python
"""2. Do zmiennej suppliers zaimportuj zawartość pliku suppliers_del_1.csv. Pobierz tylko kolumny SupplierID,
CompanyName, City"""

suppliers = pd.read_csv("/Users/kamilw/Desktop/Moje/PythonLearing/Pandas_Udemy/course-files/northwind-mongo-master/suppliers _del_1.csv",
                       usecols=['SupplierID','CompanyName','City'])
```

```python
"""3. Do zmiennej products zaimportuj zawartość pliku products_del_1.csv". Pobierz tylko kolumny ProductID,
ProductName, SupplierID**"""

products = pd.read_csv("/Users/kamilw/Desktop/Moje/PythonLearing/Pandas_Udemy/course-files/northwind-mongo-master/products_del_1.csv",
                      usecols=['ProductID','ProductName','SupplierID'])
```

```python
"""4. W zmiennej prod_and_sup zapisz wynik złączenia (inner) z products do suppliers"""

prod_and_sup = suppliers.merge(products, how='inner')
prod_and_sup.head()
```

```python
"""5. Wyświetl informację o ilości wierszy w trzech wyżej wymienionych zmiennych: suppliers, products i
prod_and_sup"""

print(len(suppliers))
print(len(products))
print(len(prod_and_sup))
```

```python
"""6. Jeżeli wszystkie dane są spójne to liczba wierszy w products powinna być równa liczbie wierszy w
prod_and_sup. Tutaj tak nie jest, więc podejrzewamy, że ktoś usunął jeden lub więcej wierszy w suppliers.
Znajdź je!
- w zmiennej prod_and_sup_outer zapisz wynik złączenia poleceniem merge dataframe products i
  suppliers (outer)
- złączając dane dodaj parametr indicator, co pozwoli określić z której tabeli pochodzi dany wiersz
- utwórz zmienną - serię wartości logicznych, która pozwoli zidentyfikować tylko wiersze, które znajdowały
  się po lewej stronie (left_only)
- wyświetl informacje o tych produktach, dla których nie jest znany dostawca"""

prod_and_sup_outer = products.merge(suppliers,how='outer',indicator=True,on='SupplierID')
prod_and_sup_outer.head()
```

```python
filter_l = prod_and_sup_outer["_merge"] == 'left_only'
```

```python
prod_and_sup_outer[filter_l]
```

```python
"""7. Podejrzewamy, że w suppliers mogą być dostawcy, którzy nie dostarczają żadnych produktów (kandydaci do
usnięcia z tego obiektu). Znajdź ich!
- korzystaj ze zmiennej prod_and_sup_outer
- utwórz zmienną - serię wartości logicznych, która pozwoli zidentyfikować tyle te wiersze, które znajdują
  się po prawej stronie (right_only)
- wyświetl informacje o tych dostawcach, dla których nie udało się znaleźć żadnych produktów"""

filter_r = prod_and_sup_outer["_merge"] == 'right_only'
```
