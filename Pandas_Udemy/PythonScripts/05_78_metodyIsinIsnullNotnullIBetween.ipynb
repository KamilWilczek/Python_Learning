```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math as math
frame = pd.read_csv('/Users/kamilw/Desktop/Moje/PythonLearing/Pandas_Udemy/course-files/Mcdonalds.csv',usecols=['Item','Category','Serving Size','Calories','TotalFat'])
```

```python
color = ['green','green','blue','blue','red','red','white']
size = ['S','S','S','M','M','M','L']
gender = ['M','W','M','W','M','W','U']
clothes = pd.DataFrame({'color':color,'size':size,'gender':gender})
clothes
```

```python
isColorBlue = clothes['color'] == 'blue'
isColorGreen = clothes['color'] == 'green'
isColorRed = clothes['color'] == 'red'
```

```python
isColorBlue
```

```python
clothes[isColorBlue | isColorGreen | isColorRed]
```

```python
# isin
# bez zmiennej, złożone
clothes[clothes["color"].isin(['blue','green','red'])]
```

```python
# isin
# ze zmienną
myColors = clothes['color'].isin(['blue','green','red'])
clothes[myColors]
```

```python
# chcemy wyświetlić pozycje mające więcej niż 300 kalorii włącznie i mniej niż 400 kalorii włącznie #1
hasMoreThan300Cal = frame["Calories"] >= 300
hasLessThan400Cal = frame['Calories'] <= 400
```

```python
frame[hasMoreThan300Cal & hasLessThan400Cal].head()
```

```python
# chcemy wyświetlić pozycje mające więcej niż 300 kalorii włącznie i mniej niż 400 kalorii włącznie #2
# deklarujemy zmienna przy użyciu metody .between()

caloriesBetween300and400 = frame['Calories'].between(300,400)
```

```python
# wykorzystanie zmienniej
frame[caloriesBetween300and400].head()
```

```python
# zagnieżdżona składnia
frame[frame['Calories'].between(300,400)].head()
```
