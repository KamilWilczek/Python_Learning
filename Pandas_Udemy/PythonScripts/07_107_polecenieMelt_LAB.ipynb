```python
"""1. Zaimportuj moduł pandas i numpy nadaj im standardowe aliasy. Zaimportuj też datetime, timedelta i time,
możesz skorzystać z poniższych poleceń:"""

import pandas as pd
import numpy as np
from datetime import datetime
from datetime import timedelta
import time
```

```python
"""2. Do wykonania zadań z wykorzystaniem polecenia melt będziemy korzystać z danych w postaci tabeli
przestawenej. Uruchom poniższy kod, który przygotuje zmienną df o odpowiedniej strukturze:
 df = pd.read_csv('./marathon_results_2016.csv', index_col='Bib',
 usecols=['Bib','40K','Half','Pace','Age','M/F','Country','State','City'])
 df['40K'] = df['40K'].apply(pd.to_timedelta)
 df['Half'] = df['Half'].apply(pd.to_timedelta)
 df['TotalSeconds'] = df['40K'].apply(lambda x: timedelta.total_seconds(x)
)
 df['HalfSeconds'] = df['Half'].apply(lambda x: timedelta.total_seconds(x)
)
 df = df.pivot_table(index="Age",columns="M/F",values="TotalSeconds").head
()
 df.head()"""

df = pd.read_csv('/Users/kamilw/Desktop/Moje/PythonLearing/Pandas_Udemy/course-files/marathon_results_2016.csv', index_col='Bib',usecols=['Bib','40K','Half','Pace','Age','M/F','Country','State','City'])
df['40K'] = df['40K'].apply(pd.to_timedelta)
df['Half'] = df['Half'].apply(pd.to_timedelta)
df['TotalSeconds'] = df['40K'].apply(lambda x: timedelta.total_seconds(x))
df['HalfSeconds'] = df['Half'].apply(lambda x: timedelta.total_seconds(x))
df = df.pivot_table(index="Age",columns="M/F",values="TotalSeconds").head()
df.head()
```

```python
"""3. Usuń indeks z obiektu df"""
df.reset_index(inplace=True)
df.head()
```

```python
"""4. Zamień dane do postaci tabeli korzystając z polecenia melt definiując kolumnę Age jako kolumnę indeksu"""

df.melt(id_vars='Age').head()
```

```python
"""5. Dodaj do poprzedniego polecenia parametr, który spowoduje, że kolumna z wartościami będzie nazwana
TotalSeconds"""

df.melt(id_vars='Age',value_name='TotalSeconds').head()
```

```python
"""6. Dodaj do poprzedniego polecenia parametr, który spowoduje, że nowo utworzona kolumna będzie miała
nagłówek Sex"""

df.melt(id_vars='Age',value_name='TotalSeconds',var_name='Sex').head()
```

```python
"""7. Podobnie jak w punkcie drugim wykonaj następujący kod, który spowoduje utworzenie nieco innego obiektu w
postaci pivot table:
df = pd.read_csv('./marathon_results_2016.csv', index_col='Bib',usecols=['Bib','40K','Half','Pace','Age','M/F','Country','State','City'])
df['40K'] = df['40K'].apply(pd.to_timedelta)
df['Half'] = df['Half'].apply(pd.to_timedelta)
df['TotalSeconds'] = df['40K'].apply(lambda x: timedelta.total_seconds(x))
df['HalfSeconds'] = df['Half'].apply(lambda x: timedelta.total_seconds(x))
df = df.pivot_table(index="Age",columns="M/F",values=["HalfSeconds","TotalSeconds"]).head()
df.head()"""

df = pd.read_csv('/Users/kamilw/Desktop/Moje/PythonLearing/Pandas_Udemy/course-files/marathon_results_2016.csv', index_col='Bib',usecols=['Bib','40K','Half','Pace','Age','M/F','Country','State','City'])
df['40K'] = df['40K'].apply(pd.to_timedelta)
df['Half'] = df['Half'].apply(pd.to_timedelta)
df['TotalSeconds'] = df['40K'].apply(lambda x: timedelta.total_seconds(x))
df['HalfSeconds'] = df['Half'].apply(lambda x: timedelta.total_seconds(x))
df = df.pivot_table(index="Age",columns="M/F",values=["HalfSeconds","TotalSeconds"]).head()
df.head()
```

```python
"""8. Usuń indeks z obiektu df"""

df.reset_index(inplace=True)
df.head()
```

```python
"""9. Zamień dane do postaci tabeli korzystając z polecenia melt definiując kolumnę Age jako kolumnę indeksu"""

df.melt(id_vars="Age").head()
```

```python
"""10. Zmień poprzednie polecenie tak, aby jawnie wskazać, że przeniesione z kolumn do wierszy (odpivotowane)"""

df.melt(id_vars='Age',value_vars=['HalfSeconds','TotalSeconds']).head()
```
